```python
import pandas as pd
import pandas_profiling

# Read New York City property sales data
df = pd.read_csv(DATA_PATH+'condos/NYC_Citywide_Rolling_Calendar_Sales.csv')

# Change column names: replace spaces with underscores
df.columns = [col.replace(' ', '_') for col in df]

# SALE_PRICE was read as strings.
# Remove symbols, convert to integer
df['SALE_PRICE'] = (
    df['SALE_PRICE']
    .str.replace('$','')
    .str.replace('-','')
    .str.replace(',','')
    .astype(int)
)
```

```python
# BOROUGH is a numeric column, but arguably should be a categorical feature,
# so convert it from a number to a string
df['BOROUGH'] = df['BOROUGH'].astype(str)
```

```python
# Reduce cardinality for NEIGHBORHOOD feature

# Get a list of the top 10 neighborhoods
top10 = df['NEIGHBORHOOD'].value_counts()[:10].index

# At locations where the neighborhood is NOT in the top 10, 
# replace the neighborhood with 'OTHER'
df.loc[~df['NEIGHBORHOOD'].isin(top10), 'NEIGHBORHOOD'] = 'OTHER'
```

Use a subset of the data where BUILDING_CLASS_CATEGORY == '01 ONE FAMILY DWELLINGS' and the sale price was more than 100 thousand and less than 2 million.

```python
df.head()
```

```python
# Creating new df 
df = df[(df['BUILDING_CLASS_CATEGORY'] == '01 ONE FAMILY DWELLINGS') & (df['SALE_PRICE'] < 2000000) & (df['SALE_PRICE'] > 100000)]
df.head()
```

Do train/test split. Use data from January â€” March 2019 to train. Use data from April 2019 to test.

```python
# Checking dtype of 'SALE_DATE'
df.dtypes
```

```python
# Changing 'SALE_DATE' from object to datetime
df['SALE_DATE'] = pd.to_datetime(df['SALE_DATE'])
```

```python
# Checking if change worked
df.dtypes
```

```python
# Create train data for Jan-March months
train = df[(df['SALE_DATE'].dt.month == 1) | (df['SALE_DATE'].dt.month == 2) | (df['SALE_DATE'].dt.month == 3)]

# Create test data for April
test = df[df['SALE_DATE'].dt.month == 4]
```

```python
# Arrange y target vectors
target = 'SALE_PRICE'
y_train = train[target]
y_test = test[target]
```

```python
# Get mean baseline
print('Mean Baseline (using 0 features)')
guess = y_train.mean()
print(guess)
```

```python
# Train Error
from sklearn.metrics import mean_absolute_error
y_pred = [guess] * len(y_train)
mae = mean_absolute_error(y_train, y_pred)
print(f'Train Error (January - March months): {mae:.2f}')
```

```python
# Test Error
y_pred = [guess] * len(y_test)
mae = mean_absolute_error(y_test, y_pred)
print(f'Test Error (April): {mae:.2f}')
```

Do one-hot encoding of categorical features

```python
# Sorting non-numeric columns by unique
train.describe(exclude='number').T.sort_values(by='unique')
```

```python
# Dropping categorical columns with high cardinality and not useful for prediction
target = 'SALE_PRICE'
not_useful = ['BUILDING_CLASS_CATEGORY', 'APARTMENT_NUMBER', 'SALE_DATE', 'LAND_SQUARE_FEET', 'ADDRESS', 'EASE-MENT']
features = train.columns.drop([target] + not_useful)
```

```python
# Creating X_train/test and y_train/test with the useful features
X_train = train[features]
y_train = train[target]
X_test = test[features]
y_test = test[target]
```

```python
# Checking the before
X_train
```

```python
import category_encoders as ce

# Using fit_transform on X_train
encoder = ce.OneHotEncoder(use_cat_names=True)
X_train = encoder.fit_transform(X_train)
```

```python
# Using encoder.transform on X_test (no .fit)
X_test = encoder.transform(X_test)
```

```python
# Checking the after
X_train
```

Do feature selection with SelectKBest

```python
# Checking how many current features
features = X_train.columns
n = len(features)
n
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.feature_selection import SelectKBest, f_regression

# Checking optimal amount of features to use (25 in this instance)
for k in range(1, len(X_train.columns)+1):
    print(f'{k} features')
    
    selector = SelectKBest(score_func=f_regression, k=k)
    X_train_selected = selector.fit_transform(X_train, y_train)
    X_test_selected = selector.transform(X_test)

    model = LinearRegression()
    model.fit(X_train_selected, y_train)
    y_pred = model.predict(X_test_selected)
    mae = mean_absolute_error(y_test, y_pred)
    print(f'Test Mean Absolute Error: ${mae:,.0f} \n')
```

```python
# Instatiating k=25
selector = SelectKBest(score_func=f_regression, k=25)

# Using fit_transform on X/y_train
X_train_selected = selector.fit_transform(X_train, y_train)
```

```python
# Using .transform on X_test (not .fit)
X_test_selected = selector.transform(X_test)
```

```python
# Checking if columns match
X_train_selected.shape, X_test_selected.shape
```

```python
# Looking at what features were selected
selected_mask = selector.get_support()
all_names = X_train.columns
selected_names = all_names[selected_mask]
unselected_names = all_names[~selected_mask]

print('Features selected:')
for name in selected_names:
    print(name)

print('\n')
print('Features not selected:')
for name in unselected_names:
    print(name)
```

Fit a ridge regression model with multiple features (Use the normalize=True parameter) and Get mean absolute error for the test set

```python
from sklearn.linear_model import Ridge
from IPython.display import display, HTML

for alpha in [0.001, 0.01, 0.1, 1.0, 1, 100.0, 1000.0]:
    
    # Fit Ridge Regression model
    display(HTML(f'Ridge Regression, with alpha={alpha}'))
    model = Ridge(alpha=alpha, normalize=True)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Get Test MAE
    mae = mean_absolute_error(y_test, y_pred)
    display(HTML(f'Test Mean Absolute Error: ${mae:,.0f}'))
```
