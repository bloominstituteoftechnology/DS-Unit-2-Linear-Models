```python
# Read New York City apartment rental listing data
import pandas as pd
df = pd.read_csv(DATA_PATH+'apartments/renthop-nyc.csv')
assert df.shape == (49352, 34)
```

```python
# Remove outliers: 
# the most extreme 1% prices,
# the most extreme .1% latitudes, &
# the most extreme .1% longitudes
df = df[(df['price'] >= 1375) & (df['price'] <= 15500) & 
        (df['latitude'] >=40.57) & (df['latitude'] < 40.99) &
        (df['longitude'] >= -74.1) & (df['longitude'] <= -73.38)]
```

# My code begins here

```python
# A bunch of imports, yes, it's probably overkill
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import math
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import plotly.express as px
```

```python
# Keep the view small
df.head(2)
```

```python
px.scatter(df, x='price', y='bedrooms', trendline='ols')
```

```python
# Instantiate this class/model/object
model = LinearRegression()
model
```

```python
# Arrange X features matrix & y target vector
features = ['bedrooms']
target = 'price'
X_train = df[features]
y_train = df[target]
```

```python
# Fit the model to train the model
model.fit(X_train, y_train)
```

```python
# Apply my findings to new data inputs
num = 3
X_test = [[num]]
y_pred = model.predict(X_test)
print(y_pred)
print(f'Predicted price of rent for a', num, 'bedroom house is $', y_pred)
```

```python
m = model.coef_
m
```

```python
b = model.intercept_
b
```

```python
# print it
print('y = m * x + b')
# print(f'y = {m:,.0f} * x + {b:,.0f}')
print('y =',m,'x +',b)
```

```python
# print(f"price = {m:,.0f} * bedrooms + {b:,.0f}")
print('price =', m,'* bedrooms +', b)
```

# Functionb & Slider It

```python
def predict(num_of_bedrooms):
  y_pred = model.predict([[num_of_bedrooms]])
  estimate = y_pred[0]
  coefficient = model.coef_[0]
  result = f'${estimate:,.0f} is the estimated price for {num_of_bedrooms:,.0f} bedrooms.'
  explained = f'Each additional bedroom adds ${coefficient:,.0f} in this model.'
  return result + '\n' + explained

print(predict(3))
```

```python
from ipywidgets import interact
interact(predict, num_of_bedrooms=(1,6));
```
