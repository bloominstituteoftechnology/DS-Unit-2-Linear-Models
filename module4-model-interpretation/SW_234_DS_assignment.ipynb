#Data Description
Personal Loan product is an unsecured loan therefore it is vital to assess the risk of the customers by checking their credit worthiness. This must be done to prevent loan defaults.

The objective is to build a Risk model using the dataset which will assess the risk of a customer defaulting after cross-selling the Personal Loan.

Column Descriptions:

Drop - index is sufficient V1: Customer ID\
v2: If a customer has bounced in first EMI (1 : Bounced, 0 : Not bounced)\
V3: Number of times bounced in recent 12 months\
V4: Maximum MOB (Month of business with TVS Credit)\
V5: Number of times bounced while repaying the loan\
V6: EMI {monthly payment}\
V7: Loan Amount\
V8: Tenure {length of loan}\
V9: Dealer codes from where customer has purchased the Two wheeler\
V10: Product code of Two wheeler (MC : Motorcycle , MO : Moped, SC : Scooter)\
V11: No of advance EMI paid\
V12: Rate of interest\
V13: Gender (Male/Female)\
V14: Employment type (HOUSEWIFE : housewife, SELF : Self-employed, SAL : Salaried, PENS : Pensioner, STUDENT : Student)\
V15: Resident type of customer\
V16: Date of birth\
V17: Age at which customer has taken the loan\
V18: Number of loans\
V19: Number of secured loans\
V20: Number of unsecured loans\
V21: Maximum amount sanctioned in the Live loans\
V22: Number of new loans in last 3 months\
V23: Total sanctioned amount in the secured Loans which are Live\
V24: Total sanctioned amount in the unsecured Loans which are Live\
V25: Maximum amount sanctioned for any Two wheeler loan\
V26: Time since last Personal loan taken (in months)\
V27: Time since first consumer durables loan taken (in months)\
V28: Number of times 30 days past due in last 6 months\
V29: Number of times 60 days past due in last 6 months\
V30: Number of times 90 days past due in last 3 months\
V31: Tier ; (Customer's geographical location)\
V32: Target variable ( 1: Defaulters / 0: Non-Defaulters)\

#Wrangle Data

```python
# Columns to drop - 2
goodbye = [
           'ID',
           'new_loans' #all zero
          ]


# Columns suspected to contain future data - 7
# Not future, information for previous loans held by clients?
future = [
          'bounce',
          'annual_bounce',
          'max_MOB',
          'total_bounce',
          'late_30_6',
          'late_60_6',
          'late_90_3'
          ]

# Columns with high % missing - 5
missing =[
          'total_credit',
          'secured_balance',
          'unsecured_balance',
          'months_last_loan',
          'months_first_loan'
          ]

# Columns to impute - 1
to_impute =['max_credit']

# Columns with missing values - 1
to_fill = ['residence']

# Date to modify, may just drop - 1
to_date =['DOB']

# Columns related to new loan - 5
loan_col= [
            'monthly_payment',
            'loan_amount',
            'installments',
            'advance_payments',
            'int_rate'
            ]

# Columns related to product - 2
prod_col = [
               'dealer',
                'product'
             ]

# Columns about the borrower - 5
person_col = [
          'sex',
          'employment',
          'Age',
          'num_unsecured_loans',
          'location'
          ]

# Skewed columns
skewed = [
              'num_secured_loans', #skewed
             'num_loans', #skewed by num secured 
          ]
```

```python
import numpy as np

# Create function to wrangle data sets

def wrangle(X):
    # copy to avoid errors
    X=X.copy()

    # DO NOT MODIFY, GREATLY IMPACTS PERMUTATION IMPORTANCES
    # Modify skewed number of secured loans column
    # X['num_secured_loans'] = np.where(X['num_secured_loans'] > 20, 100, X['num_secured_loans'])
    # X['num_loans'] = (X['num_unsecured_loans']+X['num_secured_loans'])

    # Fill in missing values in residence column
    for col in to_fill :
      X[col] = X[col].replace({np.nan:'UNKNOWN'})

    #Change DOB column to year only
    for col in to_date:
      X[col] = pd.to_datetime(X[col],infer_datetime_format=True)
      X['YOB'] =X[col].dt.year

    # Add new column, average bounces per month
    X['ave_bounce'] = ((X['total_bounce'])/(X['max_MOB']))

    # Drop extraneous columns & suspected future columns
    X = X.drop(columns = goodbye)
    # X = X.drop(columns = future)

    # Drop columns with high % missing
    X = X.drop(columns = missing)

    # Drop columns which need work
    # X = X.drop(columns = to_impute)
    # X = X.drop(columns = to_fill)
    X = X.drop(columns = to_date)
    # X = X.drop(columns = skewed)

    # Drop number of loans, because this is repetitive, and may exaggerate
    # the importance of the number of  unsecured loans
    X = X.drop(columns = 'num_loans')

    # Drop columns about new loan
    # X = X.drop(columns = loan_col)

    # Drop columns about dealer/product
    # X = X.drop(columns = prod_col)
    
    # Return wrangled dataframe
    return X
    
train = wrangle(train)
val = wrangle(val)
test = wrangle(test)

train_big = wrangle(train_big)
```

#Permutation Importances

```python
# Generate a pipeline without the model included
transformers = make_pipeline(
    ce.OrdinalEncoder(),
    SimpleImputer(strategy='mean')
)

X_train_tran = transformers.fit_transform(X_train)
X_val_tran = transformers.transform(X_val)

model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
model.fit(X_train_tran, y_train)
```

```python
# predict probabilities
yhat = model.predict_proba(X_val_tran)
# keep probabilities for the positive outcome only
yhat = yhat[:, 1]
# calculate roc curves
fpr, tpr, thresholds = roc_curve(y_val, yhat)
# plot the roc curve for the model
plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
plt.plot(fpr, tpr, marker='.', label='Logistic')
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
# show the plot
plt.show()
```

```python
# calculate the g-mean for each threshold
gmeans = (tpr * (1-fpr))*(1/2)
gmeans
```
